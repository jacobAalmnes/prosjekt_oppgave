```python
import pandas as pd
from pathlib import Path
```

```python
model_name = 'tw-roberta'

path_data_dir = Path('../data')
path_reduced = path_data_dir / Path('reduced')
path_aggregated = path_data_dir / Path('aggregated')
path_features = path_reduced / Path('features.csv')
path_final = path_data_dir / Path('final') / Path(model_name) / Path('full.csv')

path_embeddings = path_aggregated / Path(model_name) / Path('full.csv')

path_split_dir = path_data_dir / Path('split')

path_mbti = path_split_dir / Path('mbti.csv')
path_bigfive_c = path_split_dir / Path('bigfive_c.csv')
path_bigfive_s = path_split_dir / Path('bigfive_s.csv')
```

```python
df = pd.read_csv(path_features)
```

```python
df = df.drop(columns=['level_0', 'index', 'LENGTH', 'TEXT'])
df
```

```python
num_posts = df.groupby('AUTHOR').count()
num_posts
```

```python
num_posts['SOURCE']
```

```python
df = df.drop('SOURCE', axis=1)
df
```

```python
df = df.groupby('AUTHOR').mean()
```

```python
df['NUM_POSTS'] = num_posts['SOURCE']
```

```python
df
```

```python
embeddings = pd.read_csv(path_embeddings)
```

```python
final = pd.merge(df, embeddings, how='left', on=['AUTHOR', 'AUTHOR'], copy=False).set_index('AUTHOR')
```

```python
final
```

```python
final.to_csv(path_final)
```
